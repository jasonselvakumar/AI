```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from numpy import loadtxt

data = loadtxt('https://raw.githubusercontent.com/KamaleshKarthi14/Diabetes/main/diabetesclean.csv',delimiter=',') # load dataset without features
data
```

```python
data.shape
```

```python
from tensorflow import keras
```

```python
x = data[:,0:8] #Data
y = data[:,8] #Label
y
```

```python
xtrain,xtest,ytrain,ytest = train_test_split(x,y,test_size = 0.2)
```

```python
from keras.models import Sequential
from keras.layers import Dense
```

```python
model = Sequential()
model.add(Dense(8,input_dim = 8,activation='relu')) # it is input layer
model.add(Dense(8,activation='relu')) #it is a hidden layer
model.add(Dense(1,activation='sigmoid')) #it is a output layer --> softmax
```

```python
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
```

```python
model.fit(xtrain,ytrain,epochs = 350,batch_size = 43)
```

```python
_,acc = model.evaluate(xtest,ytest)
print("Accuracy for the test :",acc)
```
